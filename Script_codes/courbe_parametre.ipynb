```python
#Tests de courbe paramètres en spline pour des points quelconques au début

import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import CubicSpline

# Points donnés
x_values = [0.5, 0.35, 0.5, 0.65]
y_values = [0.25, 0.65, 0.75, 0.65]

# Créer la fonction paramétrique avec des splines cubiques
cs_x = CubicSpline(np.arange(len(x_values)), x_values)
cs_y = CubicSpline(np.arange(len(y_values)), y_values)

# Générer des valeurs t pour la courbe paramétrique
t = np.linspace(0, len(x_values) - 1, 100)

# Calculer les valeurs x et y correspondantes
x_curve = cs_x(t)
y_curve = cs_y(t)

# Tracer la courbe
plt.plot(x_curve, y_curve, label='Courbe paramétrique')
plt.scatter(x_values, y_values, color='red', label='Points donnés')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.title('Courbe paramétrique passant par les points')
plt.show()

```
